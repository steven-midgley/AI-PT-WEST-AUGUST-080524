```python
# Import our dependencies
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.utils import to_categorical 
```

```python
# Import the sample data
df = pd.read_csv("https://static.bc-edx.com/ai/ail-v-1-0/m18/lesson_1/datasets/sample-data-1000.csv")

# Plot the sample data
df.plot.scatter(x="Feature 1", y="Feature 2", c="Target", colormap="winter")
```

```python
# Separate the X and y
X = df.drop(columns="Target")
y = df["Target"]
y_binary = df["Target"]
```

```python
#MULTI-CLASS
from tensorflow.keras.utils import to_categorical 
y = to_categorical(y, num_classes=2) 
```

```python
# Use sklearn to split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
_, _, y_train, y_test = train_test_split(X, y_binary, random_state=78)
```

```python
# Create scaler instance
X_scaler = StandardScaler()

# Fit the scaler
X_scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Multiclass Classification

```python
# Create the Keras Sequential model
nn_model = tf.keras.models.Sequential()
```

```python
# Set input nodes to the number of features
input_nodes = len(X.columns)

# Add our first Dense layer, including the input layer
nn_model.add(tf.keras.layers.Dense(units=5, activation="relu", input_dim=input_nodes))
```

```python
# Add the output layer that uses a probability activation function
nn_model.add(tf.keras.layers.Dense(units=2, activation="softmax"))
```

```python
# Check the structure of the Sequential model
nn_model.summary()
```

```python
# Compile the Sequential model together and customize metrics
nn_model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

# Fit the model to the training data
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Create a DataFrame containing training history
history_df = pd.DataFrame(fit_model.history)

# Increase the index by 1 to match the number of epochs
history_df.index += 1

# Plot the loss
history_df.plot(y="loss")
```

```python
# Plot the accuracy
history_df.plot(y="accuracy")
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

## Binary Classification

```python
tf.config.list_physical_devices()
```

```python
with tf.device('CPU'):
    # Create the Keras Sequential model
    nn_model2 = tf.keras.models.Sequential()
    # Set input nodes to the number of features
    input_nodes = len(X.columns)

    # Add our first Dense layer, including the input layer
    nn_model2.add(tf.keras.layers.Dense(units=5, activation="relu", input_dim=input_nodes))
    # Add the output layer that uses a probability activation function
    nn_model2.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
nn_model2.summary()
```

```python
# Compile the Sequential model together and customize metrics
nn_model2.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python

# Fit the model to the training data
fit_model2 = nn_model2.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Create a DataFrame containing training history
history_df = pd.DataFrame(fit_model2.history)

# Increase the index by 1 to match the number of epochs
history_df.index += 1

# Plot the loss
history_df.plot(y="loss")
```

```python
# Plot the accuracy
history_df.plot(y="accuracy")
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model2.evaluate(X_test_scaled, y_test, verbose=1)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

### Binary Cross Entropy

```python
import numpy as np

def Z(x):
    return np.log(x / (1-x))
def error(Z, Y):
    return(max(Z,0) - Z*Y + np.log(1 + np.exp(-abs(Z))))

y_pred = nn_model2.predict(X_test_scaled)
y_pred = np.array(list(map(Z, y_pred))) #convert every y_pred into a log of odds value
error = np.vectorize(error)

all_errors = error(y_pred.ravel(), y_test)

np.mean(all_errors)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model2.evaluate(X_test_scaled, y_test, verbose=1)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

### Categorical Cross Entropy

```python
import numpy as np  
  
def categorical_crossentropy(Y_pred, Y_true):  
    return -np.sum(Y_true * np.log(Y_pred + 1e-9)) / Y_true.shape[0]   
  
# Assuming you have your predictions and true labels as follows:  
y_pred = nn_model.predict(X_test_scaled)  
 
  
# Ensure your predictions are in the form of probabilities and your true labels are one-hot encoded.  
# If they are not, you will have to preprocess them accordingly.  
y_test = to_categorical(y_test) 
# Calculate the categorical cross-entropy  
all_errors = categorical_crossentropy(y_pred, y_test)  
  
print(np.mean(all_errors))  
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
